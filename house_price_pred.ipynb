Simplest house price prediction

So, imagine if house pricing was as easy as a house costs 50k + 50k per bedroom, so that a 1 bedroom house costs 100k, a 2 bedroom house costs 150k etc.

How would you create a neural network that learns this relationship so that it would predict a 7 bedroom house as costing close to 400k etc.

Hint: Your network might work better if you scale the house price down. You don't have to give the answer 400...it might be better to create something that predicts the number 4, and then your answer is in the 'hundreds of thousands' etc.

```
#Importing libraries

import numpy as np
from tensorflow import keras
```

```
#Defining all functions

#Defining training data (based on problem)
def trainingData():
  xs = np.array([1, 2, 3, 4, 5])
  ys = np.array([5, 10, 15, 20, 25])
  return xs,ys

#defining model
def defineModel():
  model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
  model.compile(optimizer='sgd', loss='mean_squared_error')
  return model

#Training model
def train(model, xs, ys, epoch):
  history = model.fit(xs, ys, epochs=epoch)
  return history

#predict on new data using trained model
def predict(model, newVal):
  pred = model.predict([newVal])
  return pred
```

```
#defining number of epochs
epoch = 100  #Hyperparameter (try changing values)

#getting data
X, Y = trainingData()

#defining model
model = defineModel()

#training
history = train(model, X, Y, epoch)

#Keep an eye on loss (it should decrease)
```

```
#new prediction
val = 7.0
ans = predict(model, val)
print("Model prediction on {} is:".format(val), ans[0][0])
```
